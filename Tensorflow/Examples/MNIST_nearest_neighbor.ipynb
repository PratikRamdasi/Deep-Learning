```python
'''
Author - Pratik Ramdasi
Date - 03/26/2017
Project Title - Tensorflow MNIST practice

'''
## MNIST Classification using Tensorflow - Nearest Neighbor algorithm
# import libraries
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
```

```python
# Get input MNIST data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# Get training and testing samples
Xtr, Ytr = mnist.train.next_batch(2000)
Xte, Yte = mnist.test.next_batch(300)
```

```python
# tf graph input
xtr = tf.placeholder(tf.float32, [None, 784])
xte = tf.placeholder(tf.float32, [784])
```

```python
# Nearest neighbor calculation
distance = tf.reduce_sum(tf.abs(tf.add(xtr, tf.negative(xte))), reduction_indices=1)
pred = tf.arg_min(distance, 0)

accuracy = 0.

# initiaze variables
init = tf.global_variables_initializer()
```

```python
# run the graph
with tf.Session() as sess:
    sess.run(init)
    
    # loop over the data
    for i in range(len(Xte)):
        # get nearest neighbor
        nn_index = sess.run(pred, feed_dict={xtr: Xtr, xte: Xte[i, :]})
        # Get nearest class label and compare with true label
        print ("Test", i, "Prediction:", np.argmax(Ytr[nn_index]), "True Class:", np.argmax(Yte[i]))
        # Calculate accuracy
        if np.argmax(Ytr[nn_index]) == np.argmax(Yte[i]) :
            accuracy += 1./len(Xte)
    print "Done!"
    print ("Accuracy : ", accuracy)
```
